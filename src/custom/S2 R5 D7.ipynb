```python
from random import random, randint
from pprint import pprint
import matplotlib.pyplot as plt
from env import SquareField
```

Q learning Scenario 2

for reward 5

```python
l = 5
final_states = { (4,3):5 }
n = 1000 # number of episodes
e = 0.5 # epsilon
gamma = 0.75
alpha = 0.05

env = SquareField(l, final_states)

q_values = {}
for a in range(l):
    for b in range(l):
        for c in range(4):
            q_values[((a,b), c)] = 0

            
len_episodes = []
for i in range(n):
    current_state = env.reset()
    done = False
    next_state = None
    len_episode = 0
    reward_episode = 0
    while not done:
        action = 0
        if random()<e:
            action = randint(0,3)
        else:
            max_val = 0
            for j in range(4):
                if q_values[(current_state, j)]>max_val:
                    max_val = q_values[(current_state, j)]
                    action = j
        
        next_state, reward, done, info = env.step(action)
        max_next_state_q_value = max( [ q_values[ (next_state, j) ] for j in range(4)  ] )
        q_values[(current_state, action)] = (1-alpha)*q_values[(current_state, action)] + alpha*(reward + gamma*max_next_state_q_value)
        
        current_state = next_state
        len_episode += 1
    len_episodes.append(len_episode)
    if i%100==0:
        print(i, "Done")
```

```python
plt.plot(len_episodes)
plt.show()
```
